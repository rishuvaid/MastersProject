```python
m1=bos_jan_df['FCOUNT'].max()
m2=bos_feb_df['FCOUNT'].max()
m3=bos_mar_df['FCOUNT'].max()
m4=bos_apr_df['FCOUNT'].max()
m5=bos_may_df['FCOUNT'].max()
m6=bos_july_df['FCOUNT'].max()
m7=bos_aug_df['FCOUNT'].max()
m8=bos_sept_df['FCOUNT'].max()
m9=bos_oct_df['FCOUNT'].max()
m10=bos_nov_df['FCOUNT'].max()
m11=bos_dec_df['FCOUNT'].max()

max_data=max([m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11])

min_data=min([bos_jan_df['FCOUNT'].min(),bos_feb_df['FCOUNT'].min(),bos_mar_df['FCOUNT'].min(),bos_apr_df['FCOUNT'].min(),bos_may_df['FCOUNT'].min(),bos_july_df['FCOUNT'].min(),bos_aug_df['FCOUNT'].min(),bos_sept_df['FCOUNT'].min(),bos_oct_df['FCOUNT'].min(),bos_nov_df['FCOUNT'].min(),bos_dec_df['FCOUNT'].min()])
interval=40
thresh = [min_data + val*interval for val in range(int((max_data-min_data)/interval))]
print(thresh)
```

```python
dcmap2=folium.Map(location=[min_y,min_x])
dcmap2.geo_json(geo_path='bos_jan.geojson',key_on='feature.id',
    data=bos_jan_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_feb.geojson',key_on='feature.id',
    data=bos_feb_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_mar.geojson',key_on='feature.id',
    data=bos_mar_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_apr.geojson',key_on='feature.id',
    data=bos_apr_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_may.geojson',key_on='feature.id',
    data=bos_may_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_july.geojson',key_on='feature.id',
    data=bos_july_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_aug.geojson',key_on='feature.id',
    data=bos_aug_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_sept.geojson',key_on='feature.id',
    data=bos_sept_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_oct.geojson',key_on='feature.id',
    data=bos_oct_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_nov.geojson',key_on='feature.id',
    data=bos_nov_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)
dcmap2.geo_json(geo_path='bos_dec.geojson',key_on='feature.id',
    data=bos_dec_df,
    columns=['FID', 'FCOUNT'],
    fill_color='YlOrRd',threshold_scale=thresh)

dcmap2.lat_lng_popover()
dcmap2.create_map(path='demo_density_1.html')
```
